```python
#https://matplotlib.org/stable/gallery/text_labels_and_annotations/text_rotation.html
def addtext(ax, props):
    ax.text(0.5, 0.5, 'text 0', props, rotation=0)
    ax.text(1.5, 0.5, 'text 45', props, rotation=45)
    ax.text(2.5, 0.5, 'text 135', props, rotation=135)
    ax.text(3.5, 0.5, 'text 225', props, rotation=225)
    ax.text(4.5, 0.5, 'text -45', props, rotation=-45)
    for x in range(0, 5):
        ax.scatter(x + 0.5, 0.5, color='r', alpha=0.5)
    ax.set_yticks([0, .5, 1])
    ax.set_xticks(np.arange(0, 5.1, 0.5))
    ax.set_xlim(0, 5)
    ax.grid(True)


# the text bounding box
bbox = {'fc': '0.8', 'pad': 0}

fig, axs = plt.subplots(2, 1, sharex=True)

addtext(axs[0], {'ha': 'center', 'va': 'center', 'bbox': bbox})
axs[0].set_ylabel('center / center')

addtext(axs[1], {'ha': 'left', 'va': 'bottom', 'bbox': bbox})
axs[1].set_ylabel('left / bottom')

plt.show()

```

```python
#https://stackoverflow.com/questions/41597177/get-aspect-ratio-of-axes/42014041#42014041
from operator import sub
def get_aspect(ax):
    # Total figure size
    figW, figH = ax.get_figure().get_size_inches()
    # Axis size on figure
    _, _, w, h = ax.get_position().bounds
    # Ratio of display units
    disp_ratio = (figH * h) / (figW * w)
    # Ratio of data units
    # Negative over negative because of the order of subtraction
    data_ratio = sub(*ax.get_ylim()) / sub(*ax.get_xlim())

    return disp_ratio / data_ratio
```

```python
def find_renderer(fig):

    if hasattr(fig.canvas, "get_renderer"):
        #Some backends, such as TkAgg, have the get_renderer method, which 
        #makes this easy.
        renderer = fig.canvas.get_renderer()
    else:
        #Other backends do not have the get_renderer method, so we have a work 
        #around to find the renderer.  Print the figure to a temporary file 
        #object, and then grab the renderer that was used.
        #(I stole this trick from the matplotlib backend_bases.py 
        #print_figure() method.)
        import io
        fig.canvas.print_pdf(io.BytesIO())
        renderer = fig._cachedRenderer
    return(renderer)

from matplotlib.textpath import TextPath
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.path import Path

fig, ax1 = plt.subplots(1, 1, figsize=(8,8))


ax1.set_ylim(0 , 3)
ax1.set_xlim(0, 3)
tp1=TextPath((0.0,0.5), 'A', size=2)
#polygon=tp1.to_polygons()
#for a in polygon:
#    p1=patches.Polygon(a, fill=False)
#    ax1.add_patch(p1)
patch = PathPatch(tp1,facecolor='r',edgecolor='r',linewidth=0)
ax1.add_patch(patch)
#print(patch.get_clip_box())
bb0 = patch.get_window_extent(renderer = find_renderer(fig) )
bb = ax1.transData.inverted().transform(patch.get_window_extent(renderer = find_renderer(fig) ))
ax1.add_patch(Rectangle(bb[0],bb[1][0]-bb[0][0],bb[1][1]-bb[0][1],linewidth=2, edgecolor='b', facecolor='none'))


plt.show()
```

```python
#https://stackoverflow.com/questions/60574840/draw-linewidth-inside-rectangle-matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(8,6))

pad = 0.1 # distance between the rectangles
for i in range(3):
    for j in range(2):
        
        x = .5 + i * (1 + pad)
        y =  .5 + j * (1 + pad)
        print(i,j,x,y)
        if i == j:
            patch = patches.Rectangle((x, y), 1, 1, facecolor='blue', linewidth=0)
            tmp_path = TextPath((x,y), 'rect,lw0', size=0.15)
            text_patch = PathPatch(tmp_path,facecolor='r',edgecolor='r',linewidth=1)
        elif i < 2:
            patch = patches.Rectangle((x, y), 1, 1, facecolor='none', edgecolor='black',
                                     linewidth=10*2 if j == 0 else 10)
            if j == 0:
                tmp_path = TextPath((x,y), 'rect,lw0,clp', size=0.15)
                text_patch = PathPatch(tmp_path,facecolor='r',edgecolor='r',linewidth=1)
            else:
                tmp_path = TextPath((x,y), 'rect,lw20', size=0.15)
                text_patch = PathPatch(tmp_path,facecolor='r',edgecolor='r',linewidth=1)

        else:
            patch = patches.Ellipse((x+0.5, y+0.5), 1, 1, facecolor='none', edgecolor='black',
                                     linewidth=10*2 if j == 0 else 10)
            if j == 0:
                tmp_path = TextPath((x,y), 'ell,lw0,clp', size=0.15)
                text_patch = PathPatch(tmp_path,facecolor='r',edgecolor='r',linewidth=1)
            else:
                tmp_path = TextPath((x,y), 'ell,lw20', size=0.15)
                text_patch = PathPatch(tmp_path,facecolor='r',edgecolor='r',linewidth=1)

        ax.add_patch(patch)
        ax.add_patch(text_patch)
        if j == 0:
            patch.set_clip_path(patch)

for i in range(3):
    x = .5 + i * (1 + pad)
    for s in 0,1:
        ax.axvline (x+s, color='crimson', ls=':', lw=1)
for j in range(2):
    y =  .5 + j * (1 + pad)
    for s in 0,1:
        ax.axhline (y+s, color='crimson', ls=':', lw=1)

ax.set_xlim([0, 4.0])
ax.set_ylim([0, 3.0])
ax.set_aspect('equal')
plt.show()
```
